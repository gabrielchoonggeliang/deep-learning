```python
import numpy as np

X = np.array([1, 2, 3, 4], dtype=np.float32)
Y = np.array([3, 6, 9, 12], dtype=np.float32)

w = 0.0

# model output
def forward(x):
  return w*x

# loss = MSE
def loss(y, y_pred):
  return ((y_pred-y)**2).mean()

def gradient(x, y, y_pred):
  return np.dot(3*x, y_pred-y)

print(f'Prediction before training: f(10) = {forward(10)}')

learning_rate = 0.01
n_iters = 20

for epoch in range(n_iters):

  # dw = np.zeros_like(w)
  # print(dw)

# predict = forward pass
  y_pred = forward(X)
  print(y_pred)

  # loss
  l = loss(Y, y_pred)
  print(l)

  # calculate gradients
  dw = gradient(X, Y, y_pred)
  print(dw)

  # update weights
  w -= learning_rate * dw
  print(learning_rate)

  # output the intermediate results
  print(f'epoch {epoch+1}: w = {w:.3f}, loss = {l:.8f}')

  # dw = np.zeros_like(w)
  # print(dw) 

  # if epoch % 1 == 0:
  #   print(f'epoch {epoch+1}: w = {w:.3f}, loss = {l:.8f}')
  print()

print(f'prediction after training: f(12) = {forward(12):.3f}')
print('w is ', w)
```
